Grafique la función $f(x') = exp(-(x' - 0.5)^2 / 0.01)$ y suponga que esta representa una perturbación en una cuerda.

a) Mostrar que al hacer el cambio de variable $x' = x-ct$ la perturbación se desplaza a la derecha.

b) Mostrar que al hacer cambio de variable $x' = x+ct$ la perturbación se desplaza a la izquierda.

c) Graficar la superposición de las ondas viaqjeras de los incisos a) y b).


```python
import numpy as np
import matplotlib.pyplot as plt
```

Parámetros


```python
c = 1.0  # velocidad de propagación
t = 0.5  # tiempo
x = np.linspace(0, 1, 500)
```

Función de la perturbación  $f(x') = exp(-(x' - 0.5)^2 / 0.01)$

```python
def f(xp):
    return np.exp(-((xp - 0.5)**2) / 0.01)

f_derecha = f(x - c * t) #desplazamiento a la derecha

f_izquierda = f(x + c * t) #desplazamiento a la izquierda

f_superpuesta = f_derecha + f_izquierda #Superposición de ambas ondas

plt.figure(figsize=(12, 6)) #Graficar
```

a) Mostrar que al hacer el cambio de variable $x' = x-ct$ la perturbación se desplaza a la derecha.

```python
plt.subplot(1, 3, 1)
plt.plot(x, f_derecha, label="f(x - ct)")
plt.title("Desplazamiento a la derecha")
plt.xlabel("x")
plt.ylabel("f")
plt.grid()
plt.legend()
```

b) Mostrar que al hacer cambio de variable $x' = x+ct$ la perturbación se desplaza a la izquierda.

```python
plt.subplot(1, 3, 2)
plt.plot(x, f_izquierda, label="f(x + ct)", color="orange")
plt.title("Desplazamiento a la izquierda")
plt.xlabel("x")
plt.grid()
plt.legend()
```

c) Graficar la superposición de las ondas viaqjeras de los incisos a) y b).

```python
plt.subplot(1, 3, 3)
plt.plot(x, f_superpuesta, label="f(x - ct) + f(x + ct)", color="green")
plt.title("Superposición de ondas viajeras")
plt.xlabel("x")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
